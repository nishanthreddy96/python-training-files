```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
sam_data = pd.read_csv('sample_data.csv')
plt.plot(sam_data, color='m', linestyle='--', label = 'Ages')
plt.plot()
plt.savefig('plot.png')
```

```python
age_axis  = [21,45,65,73,76,87]
sal_axis = [0,2,4,5,6,8]
del_axis = [1,2,5,3,7,8]
sim_axis = [4,5,2,8,4,8]
plt.xlabel('Ages')
plt.ylabel('Sol Number')
plt.plot(age_axis,sal_axis, color = 'g', linestyle = '--',marker = '.', linewidth = 2,label = 'Age')
plt.plot(age_axis, del_axis, color = '#99a5e1', linestyle = ':', linewidth = 6,label = 'Del')
plt.plot(age_axis, sim_axis, color = 'm', linestyle = '-.',marker = 'o', linewidth = 1, label = 'Sim' )
plt.legend()

#plt.grid(True)
plt.style.use('seaborn-notebook')
#plt.xkcd()

plt.savefig('plot.png')
```
