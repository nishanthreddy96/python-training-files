```python
#program to choose Even Odd and Prime numbers

class find_numbers:
    
    def __init__(self,even,odd,prime,palin,fact):
        self.even = even
        self.odd = odd
        self.prime = prime
    
    def find_even(even):
        even = int(input(print("enter a number to check whether it is even or not",sep='')))
        if even%2 == 0:
            print("The number is even ")
        else:
            print("check the number and try again")
    
    def find_odd(odd):
        odd = int(input(print("enter a number to check whether it is odd or not")))
        if odd%2 !=0:
            print("the number is odd")
        else:
            print("check the number and try again")
    
    def find_prime(prime):
        prime = int(input("enter a number to check whether it is prime or not"))
        if prime > 1:
            for i in range(2,prime):
                if (prime%i) == 0:
                    print(prime, "is not a prime number")
                    print("{} times {} is {}".format(i,prime//i,prime))
                    break
            else:
                print("{} is a prime".format(prime))
        else:
            print(prime,"is not a prime number")
        
    
    def palindrome(palin):
        print("enter only word ")
        palin = input("enter a number to check palindrome or not")
        rev = ''.join(reversed(palin))
        if rev == palin:
            print("{} is palindrome".format(palin))
        else:
            print("{} is not a palindrome".format(palin))
    
    def factorial(fact):
        fact = 1
        num = int(input("enter a number to check factorial of a number"))
        for i in range(1, num+1):
            fact = fact * i
        print("The factorial of",num,"is",fact)   

        
tim = 1
num_obj = find_numbers('even','odd','prime','palin','fact')
while tim != int(0):
    print()
    print("********MENU TO CHECK A NUMBER IS EVEN OR ODD OR A PRIME NUMBER*********")
    print()
    print("\t enter 1 to check even number")
    print("\t enter 2 to check odd number")
    print("\t enter 3 to check prime number ")
    print("\t enter 4 to check palindrome ")
    print("\t enter 5 to check factorial ")
    print()
    check = int(input("choose an option from above"))
    if check == 1:
        num_obj.find_even()
    elif check == 2:
        num_obj.find_odd()
    elif check ==3:
        num_obj.find_prime()
    elif check == 4:
        num_obj.palindrome()
    elif check ==5:
        num_obj.factorial()
    elif check == 0:
        tim = 0
    else:
        print("Invalid input number...PLease try again")


```
