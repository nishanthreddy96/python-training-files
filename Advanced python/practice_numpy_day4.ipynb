```python
import numpy as np
salaries = np.genfromtxt("salary.csv",delimiter=',')
salaries
```

```python
print("The mean of above data set is\t\t\t",np.mean(salaries))
print("The median of the above dataset is\t\t",np.median(salaries))
print("The standard deviation of the above dataset is\t",np.std(salaries))
print("The variance of the above dataset is\t\t",np.var(salaries))
```

```python
angles = np.array([0,30,45,60,90])
angles_radian = angles * np.pi/180
angles_radian
```

```python
print("The sine values are\n",np.sin(angles_radian))
print()
print("The cosine values are\n",np.cos(angles_radian))
print()
print("The tan values are\n",np.tan(angles_radian))
```
