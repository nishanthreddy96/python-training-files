```python
py_ages = [43,40,55,23,37,45,46,29,12,5,7,4,10]
# Ages below 21 cannot have salary so we need to remove ages below 18
py_sal = [2100,5000,2700,7200,2190,5600,6600,4742,6765,1753,7501,3201,2301]
ar_ages = np.sort(py_ages)
ar_sal = np.sort(py_sal)
print(ar_ages)
print(ar_sal)
plt.plot(py_ages,py_sal) #secondary or 2nd comparison is assigned to orange
```

```python
# after sorting
plt.plot(ar_ages,ar_sal) # primary or 1st comparison is assigned to blue
```

```python
l_ages = []
for i in ar_ages:
    if i>21:
        l_ages.append(i)        
    else:
        pass
print(l_ages)
```

```python
l_sal = []
for i in ar_sal:
    if i> 3000:
        l_sal.append(i)
       # print(i, end = ' ')
    else:
        pass
print(l_sal)
```
